```python
# from Corey Shaefer

from matplotlib import pyplot as plt

plt.style.use("fivethirtyeight")


minutes = [1, 2, 3, 4, 5, 6, 7, 8, 9]

player1 = [8, 6, 5, 5, 4, 2, 1, 1, 0]
player2 = [0, 1, 2, 2, 2, 4, 4, 4, 4]
player3 = [0, 1, 1, 1, 2, 2, 3, 3, 4]

labels = ['player1', 'player2', 'player3']
colors = ['#6d904f', '#fc4f30', '#008fd5']

plt.stackplot(minutes, player1, player2, player3, labels=labels, colors=colors)

plt.legend(loc=(0.07, 0.05))

plt.title("My Awesome Stack Plot")
plt.tight_layout()
plt.show()

# Colors:
# Blue = #008fd5
# Red = #fc4f30
# Yellow = #e5ae37
# Green = #6d904f
```

```python
import pandas as pd
cases=pd.read_csv('MalariaCases.csv')
cases
```

```python
import matplotlib.pyplot as plt

cases.plot(x="Year", y=["World", "Africa"], kind="line")

plt.show()
```

```python
#first set of data, x and y
#make sure you have pandas and matplotlib imported
%config InlineBackend.figure_format = 'retina' # enhances the quality of the figure

#controls the size of the figure
fig = plt.figure()
fig, ax = plt.subplots(figsize=(15, 10)) #changes size of the figure

x = cases['Year']
world_y = cases['World']
plt.plot(x, world_y, color='r', linestyle='-', marker = 'o', markersize = 15, label='World')

#second set of y data
africa_y = cases['Africa']
plt.plot(x, africa_y,color='#444444', linestyle='--', marker='o', markersize = 15, label='Africa')

# add y axes as needed

plt.xlabel('Year', fontsize = 20)
plt.ylabel('Number of Cases', fontsize = 20)
plt.title('Malaria Cases 2010-2018', fontsize = 20)
plt.xticks(cases['Year'], fontsize = 16)
plt.yticks(fontsize = 16)

plt.legend(fontsize = 20)

plt.tight_layout()

plt.savefig('malariacases.pdf',dpi=1200,format='pdf',bbox_inches='tight') # same result if bbox_inches=None

plt.show()
```

```python
#first set of data, x and y
#make sure you have pandas and matplotlib imported

import numpy as np #to be able to modify axes
%config InlineBackend.figure_format = 'retina' # enhances the quality of the figure
plt.style.use('seaborn')

#controls the size of the figure
fig = plt.figure()
fig, ax = plt.subplots(figsize=(15, 10)) #changes size of the figure

x = cases['Year']
world_y = cases['World']
plt.plot(x, world_y, marker = 'o', label='World')

#second set of y data
africa_y = cases['Africa']
plt.plot(x, africa_y, marker = 'o', label='Africa')

# add y axes as needed

plt.xlabel('Year', fontsize = 20)
plt.ylabel('Number of Cases', fontsize = 20)
plt.title('Malaria Cases 2010-2018', fontsize = 20)
plt.xticks(cases['Year'], fontsize = 16)
#plt.yticks(fontsize = 16)
plt.yticks(np.arange(190000, 260000, 10000), fontsize=16)
ax.set_yticklabels(['190,000', '200,000', '210,000', '220,000', '230,000', '240,000', '250,000'])
plt.legend(fontsize = 20)

plt.tight_layout()

plt.savefig('malariacases2.pdf',dpi=1200,format='pdf',bbox_inches='tight') # same result if bbox_inches=None

plt.show()
```
