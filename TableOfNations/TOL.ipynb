# Making a stacked bar chart

```python
import numpy as np
import matplotlib.pyplot as plt
%config InlineBackend.figure_format = 'retina'

fig = plt.figure();
fig, ax = plt.subplots(figsize=(10, 10));
N = 9
Born = shem['Born']
Total = shem['Total']
ind = np.arange(N)    # the x locations for the groups
width = 0.50      # the width of the bars: can also be len(x) sequence

p1 = plt.bar(ind, Born, width,color='w');
p2 = plt.bar(ind, Total, width,
             bottom=Born,color='green');

plt.ylabel('Years', fontsize=16);
plt.title('Lifetimes of Shem and sons', fontsize=20);
plt.xticks(ind, (
'Shem',
'Arphaxahad',
'Shelah',
'Eber',
'Peleg',
'Reu',
'Serug',
'Nahor',
'Terah'
), rotation=90, fontsize=16);
plt.xlabel('Person', fontsize=16)
plt.yticks(np.arange(0, 650, 100), fontsize=16);
# plt.legend((p1[0], p2[0]), ('Born', 'Total'))
plt.savefig('lives.pdf',dpi=1200,format='pdf',bbox_inches='tight')
```

# Horizontal bar chart

```python
# pandas and numpy already imported

#plt.rcdefaults()
fig, ax = plt.subplots(figsize=(15,10))


# Enter data
person = shem['Person']
years = shem['Total']

# Plot data
plt.barh(person,years,color='g')

# Set axes
ax.set_yticks(person)
ax.set_yticklabels(person)
ax.set_ylabel('Person')
ax.invert_yaxis()  # labels read top-to-bottom
ax.set_xlabel('Years')
ax.set_title('Lifetimes of Shem and sons');
```

# Another way to make a stacked vertical bar chart

```python
# pandas and numpy already imported
# same method as above
%config InlineBackend.figure_format = 'retina'

# plt.rcdefaults();
fig, ax = plt.subplots(figsize=(10,10));


# Obtain data and arrange
person = shem['Person']
years = shem['Total']
born = shem['Born']

# Plot data
p1=plt.bar(person,born,color='w')
p2=plt.bar(person,years,bottom=born)

# Set axes
plt.xticks(person,rotation=90,fontsize=12)
# plt.yticks([0,100,200,300,400,500,600,700]), or alternatively,
plt.yticks(np.arange(0, 650, 100),fontsize=12);
plt.xlabel('Person',fontsize=12)
plt.ylabel('Years',fontsize=12)
plt.title('Lifetimes of Shem and sons',fontsize=16);
```
