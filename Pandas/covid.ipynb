```python
#Worldwide locations
import matplotlib.pyplot as plt

# x axis values 
x = cov['Country/Region'];
# corresponding y axis values 
y = cov['Deaths'];
# x tick labels; used in plt.xticks() below
labels = cov['Country/Region']; 

# plotting the graph
fig = plt.figure();
fig, ax = plt.subplots(figsize=(30, 10)); #changes size of the figure
# plotting line and scatter superimposed
plt.bar(x,y,color='k');
plt.xlabel('Country');
plt.ylabel('Deaths');
plt.title('Deaths by Country Where Cases Reported')
plt.xticks(x,labels,rotation=90); # substitutes x tick labels for x numbers
```

```python
uslocs=cov.loc[cov['Country/Region']=='US']
uslocs
```

```python
#US locations
import numpy as np

# x axis values 
x = uslocs['Province/State']
# corresponding y axis values 
y = uslocs['Deaths']
# x tick labels; used in plt.xticks() below
labels = uslocs['Province/State'] 
ylabels = uslocs['Deaths']

# plotting the graph
fig = plt.figure();
fig, ax = plt.subplots(figsize=(20, 10)); #changes size of the figure
# plotting line and scatter superimposed
plt.bar(x,y,color='b');
plt.xlabel('County and State',fontsize=16);
plt.ylabel('Deaths',fontsize=16);
plt.title('Deaths by US Location Where Cases Reported',fontsize=16)
plt.xticks(x,labels,rotation=90); # substitutes x tick labels for x numbers
plt.yticks(np.arange(0,11,step=1));# sets the parameters of yticks: start,end (exclusive), steps
```
